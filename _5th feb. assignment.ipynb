```python
#QUESTION 1:Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.
#ANSWER:
#In Object-Oriented Programming (OOP), a class is a blueprint for creating objects,
#providing initial values for state (variables or attributes),
#and implementations of behavior (member functions or methods). An object is an instance of a class. When the individual objects are created,
#they inherit all the variables and methods from the class.For example, consider a class named Car. A car has attributes such as weight and color,
#and methods such as drive and brake. We can define a class named Car that would have attributes like weight and color and methods like drive and brake.
#We can then create any number of objects from this class, each representing a different car with its own unique attributes.
```

```python
#Q2. Name the four pillars of OOPs.
#ANSWER:
#The four pillars of Object-Oriented Programming (OOP) in Python (as well as in other languages) are:

#1. Encapsulation: Encapsulation refers to the practice of containing the properties and behaviors of a class within the class itself. 
#This means that the internal workings of a class are hidden from outside code, and can only be accessed through a well-defined interface.
#2. Inheritance: Inheritance is a hierarchical mechanism that allows a new class to be based on an existing class and extend or modify its properties and/or behaviors as needed.
#This can save a lot of time and effort in code development, as existing code can be reused and modified as needed.
#3. Polymorphism: Polymorphism refers to the ability of a class to take on multiple forms,
#based on the context in which it is used. In Python, this is usually achieved through the use of inheritance and method overriding.
#4. Abstraction: Abstraction refers to the practice of modeling complex systems or concepts in simplified ways. In the context of OOP,
#this means defining abstract classes that provide a high-level view of a system or concept, without specifying all the implementation details.
```

```python
#Q3. Explain why the __init__() function is used. Give a suitable example.

#ANSWER:
#The init() function is a special method in Python that is automatically called when an instance (object) of a class is created.
#It is primarily used to initialize the properties (attributes) of the instance with specific values or states.

#For example, consider a class Person that has two attributes name and age.
#To initialize these attributes when an instance of the class is created, we can define the init() function in the class as follows:

python
class Person:
    def init(self, name, age):
        self.name = name
        self.age = age


#In this code snippet, the init() function initializes the name and age attributes of a Person object using two parameters passed to it (name and age). 
#The self parameter is a reference to the instance being created, and is automatically passed to the function when it is called.

#Now, when we create an instance of the Person class and pass values for name and age, the init() function is automatically called and the attributes of the instance are initialized:

python
person1 = Person("John", 30)
print(person1.name)  # Output: John
print(person1.age)  # Output: 30


#Here, we create a new Person object person1 with the name "John" and age 30. The init() function initializes the name and age attributes of the object person1, which we can then access and print using dot notation.
```

```python
#Q4. Why self is used in OOPs?
#ANSWER:
#In object-oriented programming (OOP) in Python, self is a parameter that refers to the instance of the class that is being created or manipulated.
#It is a convention followed in Python, and typically the first parameter of any instance method.
#The purpose of self is to allow instance methods to access and modify the attributes of the instance,
#without conflicting with other instances of the same class. By using self, methods can reference instance variables and methods within the class.

#For example, consider a class Person with the attributes name and age:

python
class Person:
    def init(self, name, age):
        self.name = name
        self.age = age

    def describe(self):
        return f"{self.name} is {self.age} years old."


#In this code, the describe() method returns a string that describes the Person instance's name and age attributes. 
#self is used to access the instance attributes directly. If self was not used, the method would have no way of accessing the instance attributes.

#Overall, self is an essential part of OOP in Python,
#as it allows classes to maintain state and interact with instance attributes in a logical and organized manner.
```

```python
#Q5. What is inheritance? Give an example for each type of inheritance.

#ANSWER:
#Inheritance is a fundamental concept in object-oriented programming that allows a class to inherit the properties (attributes and methods) of another class. The class which is being inherited from is called the superclass or parent class, and the class which inherits from it is called the subclass or child class. 
#There are four types of inheritance in Python:
#1. Single Inheritance: In single inheritance, a subclass inherits from a single parent class. 
#Example:

python
class Animal:
    def init(self, name):
        self.name = name
        
    def speak(self):
        pass  # abstract method


class Dog(Animal):
    def speak(self):
        return "Woof!"


d = Dog("Buddy")
print(d.name)  # "Buddy"
print(d.speak())  # "Woof!"

#In this example, the Animal class has an abstract speak method that returns nothing. The Dog class inherits from the Animal class and overrides the abstract method to return "Woof!".
#2. Multiple Inheritance: In multiple inheritance, a subclass inherits from multiple parent classes.

Example:

python
class A:
    def method(self):
        return "A"


class B:
    def method(self):
        return "B"


class C(A, B):
    pass


c = C()
print(c.method())  # "A"


#In this example, both A and B classes have a method named method.
#The C class inherits from both classes, but since it inherits from A first, calling c.method() returns "A".

#3. Multilevel Inheritance: This is when a subclass inherits from a parent class which in turn inherits from another parent class.

Example:

python
class A:
    def method(self):
        return "A"


class B(A):
    pass


class C(B):
    pass


c = C()
print(c.method())  # "A"


#In this example, the C class inherits from the B class which in turn inherits from the A class.

#4. Hierarchical Inheritance: This is when multiple subclasses inherit from a single parent class.

Example:

python
class A:
    def method(self):
        return "A"


class B(A):
    pass


class C(A):
    pass


b = B()
c = C()
print(b.method())  # "A"
print(c.method())  # "A"


#In this example, both B and C classes inherit from the A class. Both objects b and c have a method that returns "A".
```
